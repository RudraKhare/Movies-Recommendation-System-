```python
#Since genres is in format of string of list we have to convert it to list
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
# import json

# movies['genres'] = movies['genres'].apply(lambda x: json.loads(x))
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
#WE want only name of 3 actors. so we will write a funtion that will extract first three actor names
def convert3(obj):
  L=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter != 3:
      L.append(i['name'])
      counter+=1
    else:
      break
  return L
```

```python
movies['cast']= movies['cast'].apply(convert3)
```

```python
movies['crew']
```

```python
def fetch_director(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']== 'Director':
      L.append(i['name'])
      break
  return L
```

```python
movies['crew']= movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
#Overview is in string format wo we will convert it to an list format so that we can concatenate it with other list
movies['overview']=movies['overview'].apply(lambda x : x.split())
```

```python
#Removing Space between words
movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

Stemming

```python
# We are applying steming over here (reason is described below )
!pip install nltk
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():#Convert String to list
    y.append(ps.stem(i)) # Steming Every Word In The List and the appending it to y
  return " ".join(y) # Again converting it back to string
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'].apply(lambda x:x.lower() )
```

Vectorization

```python
# we will create a 2d space where each movie will have its vector plotted in that space(there will be 5000 vectors for each movie) and nearest vector to it will be given as recommend movies
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
#we converted it to numpy array beacause it return an object which is in form of sparse matrix
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```
